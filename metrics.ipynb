```python
from sklearn.metrics import confusion_matrix    #CONFUSION METIX


y_true =[0, 1, 0, 1, 1, 0, 1, 1]
y_pred =[0,1, 1, 0, 1, 0, 1, 1]

mout= confusion_matrix(y_true , y_pred)
print(mout)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score, f1_score   #CLASSIFICATION METRICS


y_true =[0, 1, 0, 1, 1, 0, 1, 1]
y_pred =[0,1, 1, 0, 1, 0, 1, 1]

print("Accuracy: ", accuracy_score(y_true , y_pred))
print("Precision : ", precision_score(y_true,y_pred))
print("recall : ", recall_score(y_true,y_pred))
print("f1 score  : ", f1_score(y_true,y_pred))

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np 

real_marks = [90, 100, 80, 95]
pred_marks = [85, 95, 70, 90]

mae = mean_absolute_error(real_marks, pred_marks)
mse=  mean_squared_error(real_marks, pred_marks)
rmse = np.sqrt(mse)
print("MAE :" , mae)
print("MSE : ", mse)
print("RMSE : ", rmse)
```
