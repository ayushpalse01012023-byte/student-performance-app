```python
import pandas as pd 
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt 

df = pd.DataFrame({
    "hours": [2, 3, 2.5, 20, 21, 50],
    "marks": [40, 45, 42, 80, 82, 30]
})

X= df[["hours", "marks"]]
model = DBSCAN(eps= 5, min_samples=2 )
label = model.fit_predict(X)

plt.scatter(df["hours"], df["marks"], c=label, cmap = "viridis")
plt.xlabel("hours")
plt.ylabel("marks")
plt.legend()
plt.grid(True)
plt.show()
label

```
