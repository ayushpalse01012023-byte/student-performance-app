```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = {
    'name': ['ayush','piyush','ritesh','hitesh', 'mitesh','rushi'],
    'age': [20, 30, 40, 22, 38, 25],
    'spend': [100, 200, 300, 120, 250, 140]

}
df = pd.DataFrame(data)
X = df[['age', 'spend']]
model = KMeans(n_clusters = 2, random_state=42, n_init=10)#KMEANS FIND THE PATTERN AND GROUP THE DATA MAKE THE CLUSTERS 
df['groups'] = model.fit_predict(X)
plt.figure(figsize=(6,7))
for group in df['groups'].unique():#0,1
    group_data = df[df['groups'] == group]
    plt.scatter(group_data['age'], group_data['spend'], label = f"Group {group}")

plt.xlabel("age")
plt.ylabel("spend")
plt.legend()
plt.grid(True)
plt.show()

df
```

```python
import pandas as pd 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = {
    'name': ['A', 'B', 'C', 'D', 'E', 'F'],
    'hours': [2, 3, 8, 9, 4, 10],
    'marks': [40, 45, 90, 95, 50, 98]
}
df = pd.DataFrame(data)
X= df[['hours', 'marks']]
model = KMeans(n_clusters = 2, random_state=42, n_init=10)
df["grouped"] = model.fit_predict(X)

for group in df["grouped"].unique():
    group_data = df[df["grouped"] == group]
    plt.scatter(group_data["hours"], group_data["marks"], label=f"group {group}")
plt.xlabel("Hours")
plt.ylabel("marks")
plt.grid(True)
plt.legend()
plt.show()
df
```

```python
import pandas as pd 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = {
    'name': ['A', 'B', 'C', 'D', 'E', 'F'],
    'experience': [1, 2, 3, 10, 12, 15],
    'salary': [25000, 30000, 35000, 80000, 90000, 100000]
}
df = pd.DataFrame(data)

X= df[["experience", "salary"]]
model = KMeans(n_clusters = 2, n_init = 10, random_state=42)

df["group"]= model.fit_predict(X)

for group in df["group"].unique():
    group_data = df[df["group"] == group]
    plt.scatter(group_data["experience"], group_data["salary"], label= f"Group {group}")
plt.xlabel("EXPERIENCE")
plt.ylabel("SALARY")
plt.grid(True)
plt.legend()
plt.show()
df
```

```python
import pandas as pd 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = {
    'house': ['H1', 'H2', 'H3', 'H4', 'H5', 'H6'],
    'area': [500, 600, 650, 1500, 1600, 1700],
    'price': [20, 25, 28, 80, 85, 90]
}
df = pd.DataFrame(data)
X =df[["area","price"]]

model = KMeans(n_clusters= 2, n_init = 10, random_state=42)
df["group"]= model.fit_predict(X)
for group in df["group"].unique():
    group_data = df[df["group"] == group]
    plt.scatter(group_data["area"], group_data["price"], label = f"group : {group}")
plt.xlabel("Area")
plt.ylabel("PRICE")
plt.grid(True)
plt.legend()
plt.title("KMEANS")
plt.show()
df
```

```python
from sklearn.cluster import DBSCAN
import pandas as pd 
import matplotlib.pyplot as plt

import pandas as pd

df = pd.DataFrame({
    "x": [1, 2, 2, 3, 20, 21, 22, 23, 50],
    "y": [1, 1, 2, 2, 20, 21, 20, 22, 10]
})

X= df[["x", "y"]]
model = DBSCAN(eps = 4, min_samples =2)
df["cluster"]= model.fit_predict(X)
df
for group in df["cluster"].unique():
    grouped = df[df["cluster"] == group]
    plt.scatter(grouped["x"], grouped["y"],color = "blue")

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
df
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt



df = pd.DataFrame({
    "x": [1,2,2,3, 20,21,22,23, 40,41,42,43],
    "y": [1,1,2,2, 20,21,20,22, 5,6,5,7]
})
X= df[["x", "y"]]
std = StandardScaler()
scaled = std.fit_transform(X)
model = DBSCAN(eps = 0.3, min_samples = 2)
df["group"] = model.fit_predict(scaled)
plt.scatter(df["x"], df["y"], c=df["group"], cmap="viridis")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
df


```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({ 
    "x": [1,2,2,20,21,50,70,90],
    "y": [1,1,2,20,21,10,80,5]
})
X = df[["x", "y"]]
std = StandardScaler()
scaled = std.fit_transform(X)

model = DBSCAN(eps = 0.4, min_samples = 2)
df["group"]= model.fit_predict(scaled)
plt.scatter(df["x"], df["y"], c=df["group"], cmap="viridis")
plt.plot(df["x"], df["y"])
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
df
```
