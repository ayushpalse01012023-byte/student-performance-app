```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 

data = {
    "age": [25, 30, 35, 40, 45, 50],
    "income": [30000, 40000, 50000, 60000, 70000, 80000],
    "spendings" :[70, 60, 50, 40, 30, 20],
    "savings": [1000, 5000,8000, 10000, 15000,20000 ]
}
df = pd.DataFrame(data)
std = StandardScaler()
std_scaled = std.fit_transform(df)
model = PCA(n_components=2)
model_ft = model.fit_transform(std_scaled)
model_df= pd.DataFrame(model_ft, columns = ["PCA1","PCA2"])
explained_variance = model.explained_variance_ratio_
print("Variaance of the each component : ")
print(np.round(explained_variance *100, 2))

plt.figure(figsize=(6,8))
plt.scatter(model_df["PCA1"], model_df["PCA2"], color= "black")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("PCA components")
plt.grid(True)
plt.show()
model_df 
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

data = {
    'Math': [90, 85, 88, 40, 45, 50],
    'Science': [92, 86, 89, 42, 48, 52],
    'English': [88, 82, 85, 38, 44, 48],
    'Computer': [95, 90, 92, 50, 55, 60]
}
df= pd.DataFrame(data)
std = StandardScaler()
scaled = std.fit_transform(df)
model = PCA(n_components= 2)
model_ft = model.fit_transform(scaled)
model_df = pd.DataFrame(model_ft, columns=["col1", "col2"])

explain = model.explained_variance_ratio_
print("Ratio : ")
print(np.round(explain*100, 2))
plt.figure(figsize= (8,9))
plt.scatter(model_df["col1"], model_df["col2"], color= "red")
plt.xlabel("COL1")
plt.ylabel("COL2")
plt.title("PCA")
plt.grid(True)
plt.legend()
plt.show()
model_df

```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

data = {
    'Screen_Time': [2, 3, 2.5, 8, 9, 7.5],
    'Data_Usage': [1, 1.5, 1.2, 5, 6, 5.5],
    'Apps_Used': [10, 12, 11, 35, 40, 38],
    'Battery_Usage': [20, 25, 22, 70, 80, 75]
}
df = pd.DataFrame(data)
std = StandardScaler()
scaled = std.fit_transform(df)

model = PCA(n_components =2)
model_ft = model.fit_transform(scaled)
model_df = pd.DataFrame(model_ft, columns =["col1", "col2"])

explained = model.explained_variance_ratio_
print("Ratio : ")
print(np.round(explained*100, 2))
plt.figure(figsize= (8,10))
plt.scatter(model_df["col1"], model_df["col2"], color="red")
plt.xlabel("col1")
plt.ylabel("col2")
plt.title("components")
plt.grid(True)
plt.legend()
plt.show()
model_df
```
