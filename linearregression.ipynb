```python
from sklearn.linear_model import LinearRegression

X = [[1],[2],[3],[4],[5],[6]]
y= [10000,20000, 30000, 40000, 50000, 60000]

model = LinearRegression()
model.fit(X,y)
exp = float(input("Enter your experience : "))
salary = model.predict([[exp]])

print(f"The salary with your {exp}experience will be {salary} rupees")
```

```python
from sklearn.linear_model import LinearRegression

X= [
    [1,6],[2,6],[3,5],[4,4],[5,2],[6,1]
]
y =[20, 30, 40, 50, 60, 70]

model = LinearRegression()

model.fit(X,y)

studyhours = float(input("Enter study hours : "))
sleephours = float(input("Enter the sleep hours : "))

predicted_marks = model.predict([[studyhours, sleephours ]])
print(f"The predicted marks is : {predicted_marks}")
```
