```python
from sklearn.cluster import AgglomerativeClustering #hierarchial clustering finds the similar pattern in the cluster and create a new cluster [reduces the cluster ]
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler


df = pd.DataFrame({
    "x":[1,2,20,21],
    "y":[1,2,20,21]
})

X = df[["x", "y"]]
std = StandardScaler()
scaled = std.fit_transform(X)
model = AgglomerativeClustering(n_clusters=2)
label = model.fit_predict(scaled)
plt.scatter(df["x"], df["y"], c=label, cmap="viridis")
plt.plot(df["x"], df["y"])
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
label
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import pandas as pd 
df = pd.DataFrame({
    "x":[1,2,2,3,20,21,22,23,40,41,42,43],
    "y":[1,1,2,2,20,21,20,22,5,6,5,7]
})
X = df[["x", "y"]]
model = AgglomerativeClustering(n_clusters =3 )
label= model.fit_predict(X)

plt.scatter(df["x"], df["y"], c=label,  cmap="viridis")
plt.plot(df["x"], df["y"])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
