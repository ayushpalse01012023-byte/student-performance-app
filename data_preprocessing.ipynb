```python
import pandas as pd

data= {
    "name": ["ayush", "pratik", "piyush","soham", "nishant"],
    "age": [None , 13, 15, None, 18],
    "grades": [80, None, 89, 78, None]
}
df= pd.DataFrame(data)
print(df)
df["age"].fillna(df["age"].mean(), inplace= True )
df["grades"].fillna(df["grades"].mean(), inplace= True)
df 
```

```python
nullvalue = df.isnull().sum()
print(nullvalue)
```

```python
drop_na= df.dropna()
print(drop_na)
```
