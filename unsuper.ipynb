```python
import numpy as np
#Euclidean Distance 
p1 = np.array([20, 100])
p2 = np.array([30, 200])
p3 = np.array([40, 500])

distance = np.linalg.norm(p2 - p1)
dist = np.linalg.norm(p3 -p1)

print(distance)
print(dist)
```

```python
import numpy as np
#Variance and the PCA will used the second array because it has the high variance 
arr1 = np.array([100, 105, 110, 115, 120, 125])
arr2 = np.array([10, 100, 200, 300, 400, 500])

print(np.var(arr1))
print(np.var(arr2))
```
